```python
# Größen definieren
span_positions = [0, 4.9, 7.6, 10]
chord_lengths = [0.945, 0.76, 0.6, 0.4]
offsets = [0, 0.006, 0.039, 0.097]
twists = [0,]*4
airfoils = ['test']*4

# Flügel Objekt erzeugen
from wingstructure import geometry

wing = geometry.Wing.create_from_planform(span_positions, chord_lengths, offsets, twists, airfoils)
wing
```

```python
from scipy import interpolate
import numpy as np

x = [0, 4.9, 7.6, 10]
y = [0.945, 0.76, 0.6, 0.4]
n = int(round(wing.aspect_ratio())*4-1)
calculation_positions = wing.span_width()/2 * np.cos(np.arange(1, n+1)*np.pi/(n+1))

interpolate.interp1d(x,y)(np.abs(calculation_positions))
```
