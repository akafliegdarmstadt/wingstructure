# Examples for geometry module functionality

## Airfoil interpolation

```python
%matplotlib inline

coords = np.loadtxt('ah93157.dat', skiprows=1)
coords2 = np.loadtxt('naca2412.dat', skiprows=1)
airfoil1 = geometry.Airfoil(coords)
airfoil2 = geometry.Airfoil(coords2)

airfoil3 = airfoil1.interpolate(airfoil2,0.5)

airfoil1.plot('r--')
airfoil2.plot('b--')
airfoil3.plot('g')
```

## Definition of wing geometry

### Definition of planform.

```python
from wingstructure import geometry

# create shortcut
Point = geometry.Point

# create new WingExt Object
wing = geometry.WingExt(pos = (2,1)) # pos defines offset for wing coordinate system

# create first wing section
tmp_sec = geometry.Section(pos = Point(x=0,y=0,z=0), chord = 0.7, alpha = np.deg2rad(2), airfoil='NACA0012')
# add section to wing
wing.add_section(tmp_sec)

# create further sections and add them
wing.add_section(geometry.Section(Point(0.1, 4,0), 0.55, np.deg2rad(1), 'NACA0012'))
wing.add_section(geometry.Section(Point(1., 7,0), 0.35, np.deg2rad(1), 'NACA0012'))

wing.plot()
```

### Add flaps and airbrakes

```python
# set airbrake position
wing.set_airbrake(span_pos_start=1, span_pos_end=3)

# add flap
wing.set_flap(name='QR', span_pos_start=1, span_pos_end=7, depth=2*[0.25])

wing.plot()
```

__Warning:__ As seen above, the module does not check if flap is possible
