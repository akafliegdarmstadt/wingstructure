```python
import numpy as np
import collections as col
from wingstructure import structure
```

# Define materials

```python
from wingstructure.structure.material import IsotropicMaterial
Material = col.namedtuple('Material', ['ρ'])

fabric = IsotropicMaterial(ρ=0.3e3, E=210e3, G=0.0)
foam = IsotropicMaterial(ρ=0.18e3, E=210e3, G=0.0)
```

# Define structure

## import dat file & create SectionBase

```python
# load airfoil.dat file
coords = np.loadtxt('airfoils/ah93157.dat', skiprows=1)*1.2

# initialize base for section analysis
secbase = structure.SectionBase(coords)
secbase
```

## add structure elements

**start with three shell layers**

```python
out = structure.Layer(fabric, 2e-3)
core = structure.Layer(fabric, 5e-3)
inner = structure.Layer(fabric, 2e-3)
kohla = structure.Reinforcement(fabric, 15e-3, limits=[0.5, 0.6])
secbase.extend((out, core, inner, kohla))
secbase
```

**add an I-spar**

```python
spar = structure.ISpar({'flange': fabric, 'web': fabric}, 0.5, 0.3, 3e-2, 0.8, 5e-3)
secbase.append(spar)
secbase
```

**replace I-spar with boxspar**

```python
secbase.pop()
```

```python
secbase
```

```python
boxspar = structure.BoxSpar(fabric, 0.5, 0.2, 0.03, 0.01)
secbase.append(boxspar)
secbase
```

# Analyse mass

```python
with open('test.svg','w') as afile:
    afile.write(secbase._repr_svg_())
```
